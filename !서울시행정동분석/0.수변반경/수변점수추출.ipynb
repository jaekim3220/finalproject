```python
from pandas import read_excel, DataFrame
```

```python
origin = read_excel("수변_반경별_행정동.xlsx")
origin.head()
```

```python
all_value_counts = origin.stack().value_counts().sort_index().to_dict()
```

```python
all_value_counts
```

```python
df = DataFrame(list(all_value_counts.items()), columns=["읍면동명", "수변점수"])
df.to_excel("수변점수.xlsx",index=False)
```
