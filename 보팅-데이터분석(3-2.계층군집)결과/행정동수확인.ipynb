# excel 파일을 병합해 행정동의 수를 추출

```python
from pandas import read_excel, concat, DataFrame
```

```python
origin = read_excel("그린지수_사회적약자인프라밀집도_voting.xlsx")
origin1 = read_excel("그린지수_여가생활지수_voting.xlsx")
origin2 = read_excel("그린지수_접근성지수_voting.xlsx")
origin3 = read_excel("사회적약자인프라밀집도_접근성지수_voting.xlsx")
origin4 = read_excel("여가생활지수_사회적약자인프라밀집도_voting.xlsx")
origin5 = read_excel("여가생활지수_접근성지수_voting.xlsx")
```

```python
tmp = concat([origin, origin1, origin2, origin3, origin4, origin5])
tmp
```

```python
tmp1 = DataFrame(tmp['읍면동명'].value_counts())
tmp1
```

```python
tmp1.to_excel("읍면동명Voting.xlsx")
```
